```python
#| export
def download_paddleocr_models(lang: str = 'en') -> None:
    from paddleocr import PaddleOCR
    PaddleOCR(
        use_doc_orientation_classify=False,
        use_doc_unwarping=False,
        use_textline_orientation=True,
        lang=lang,
    )
```
